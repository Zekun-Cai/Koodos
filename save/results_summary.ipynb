```python
import os
import re
import pandas as pd
import numpy as np
```

```python
root_dir = './'
data = {}

for folder_name in os.listdir(root_dir):
    match = re.match(r"pred_(\w+)_\d+", folder_name)
    
    if match:
        dataset_name = match.group(1)
        score_file_path = os.path.join(root_dir, folder_name, "scores.txt")

        with open(score_file_path, 'r') as file:
            first_line = file.readline().strip()
            score_match = re.search(r"All Metric ([\d\.]+)", first_line)
            score = float(score_match.group(1))
            
            if dataset_name not in data:
                data[dataset_name] = []
            data[dataset_name].append(score)
            
df = pd.DataFrame.from_dict(data, orient='index')
df['Mean'] = df.mean(axis=1)
df['Std'] = df.std(axis=1)

df['Error (%)'] = (1 - df['Mean']) * 100
df['Std (%)'] = df['Std'] * 100

df.loc[['Twitter', 'Cyclone', 'House'], ['Error (%)', 'Std (%)']] = np.nan

df = df.reindex(['Moons', 'MNIST', 'Twitter', 'YearBook', 'Cyclone', 'House'])
df
```
